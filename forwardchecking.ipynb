```python
import pandas as pd
import numpy as np
import time
start_time = time.time()


def find_empty_cell(board):
    for row in range(9):
        for col in range(9):
            if board[row][col] == 0:
                return row, col
    return None

def is_valid_move(board, row, col, num):
    # Check the row
    if num in board[row]:
        return False
    
    # Check the column
    if num in [board[i][col] for i in range(9)]:
        return False
    
    # Check the 3x3 box
    box_row, box_col = row // 3 * 3, col // 3 * 3
    for i in range(box_row, box_row + 3):
        for j in range(box_col, box_col + 3):
            if board[i][j] == num:
                return False
    
    return True

def update_legal_values(board, row, col, num, legal_values):
    for i in range(9):
        if i != col and board[row][i] == 0:
            legal_values[row][i].discard(num)
        
        if i != row and board[i][col] == 0:
            legal_values[i][col].discard(num)
    
    box_row, box_col = row // 3 * 3, col // 3 * 3
    for i in range(box_row, box_row + 3):
        for j in range(box_col, box_col + 3):
            if (i != row or j != col) and board[i][j] == 0:
                legal_values[i][j].discard(num)

def initialize_legal_values(board):
    legal_values = [[set(range(1, 10)) for _ in range(9)] for _ in range(9)]
    for row in range(9):
        for col in range(9):
            if board[row][col] != 0:
                update_legal_values(board, row, col, board[row][col], legal_values)
    return legal_values

def forward_checking_sudoku(board):
    empty_cell = find_empty_cell(board)
    
    if not empty_cell:
        return True  # No empty cells, the puzzle is solved
    
    row, col = empty_cell
    legal_values = initialize_legal_values(board)

    for num in legal_values[row][col]:
        if is_valid_move(board, row, col, num):
            board[row][col] = num
            update_legal_values(board, row, col, num, legal_values)

            if forward_checking_sudoku(board):
                return True

            # If this value doesn't lead to a solution, backtrack
            board[row][col] = 0
    
    return False

# Example Sudoku puzzle (0 represents empty cells)
if __name__ == "__main__":
    # Load the Kaggle dataset (replace 'sudoku_dataset.csv' with your dataset file)
    dataset = pd.read_csv('sudoku.csv')

    for index, row in dataset.iterrows():
        puzzle = np.array(list(map(int, list(row['puzzle'])))).reshape(9, 9)
        
        if forward_checking_sudoku(puzzle):
           print(f"Solved Puzzle {index + 1}:")
           for row in puzzle:
              print(row)
        else:
           print(f"No solution found for Puzzle {index + 1}.")
        print("=" * 30)
        if index == 3000:
            break
end_time = time.time()
print(start_time)
print(end_time)


```
